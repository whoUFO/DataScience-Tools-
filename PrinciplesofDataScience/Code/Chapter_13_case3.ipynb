```python
from sklearn import datasets,metrics
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
iris = datasets.load_iris()
```

```python
type(iris)
```

```python
x_train,x_test,y_train,y_test=train_test_split(iris.data,iris.target)
```

```python
feature_columns=[tf.contrib.layers.real_valued_column("",dimension=4)]
optimizer= tf.train.GradientDescentOptimizer(learning_rate=.1)
classifier=tf.contrib.learn.LinearClassifier(feature_columns=feature_columns,optimizer=optimizer,n_classes=3)
classifier.fit(x=x_train,y=y_train,steps=2000)
```

```python
accuracy_score= classifier.evaluate(x=x_test,y=y_test)
```

```python
print('Accuracy:{0:f}'.format(accuracy_score["accuracy"]))
```

```python
# 使用线性回归模型进行预测
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
logreg.fit(x_train,y_train)
y_pred=logreg.predict(x_test)
accuracy=metrics.accuracy_score(y_pred,y_test)
print("accuracy:{}".format(accuracy))
```

```python
feature_columns = [tf.contrib.layers.real_valued_column("", dimension=4)]
optimizer = tf.train.GradientDescentOptimizer(learning_rate=.1)
classifier = tf.contrib.learn.DNNClassifier(feature_columns=feature_columns,
                                            hidden_units=[10, 20, 10],
                                            optimizer=optimizer,
                                            n_classes=3)
classifier.fit(x=x_train, y=y_train, steps=2000)
```

```python
accuracy_score= classifier.evaluate(x=x_test,y=y_test)
```
